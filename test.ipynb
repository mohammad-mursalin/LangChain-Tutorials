```python
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
import os

load_dotenv()  # loads GEMINI_API_KEY

# Create a Gemini 2.0 Flash model
geminiChatModel = ChatGoogleGenerativeAI(
    model="gemini-2.0-flash",
    api_key=os.getenv("GEMINI_API_KEY")
)

identities = """
5. (a - b + c)2 = a2 + b2 + c2 - 2ab + 2ac - 2bc
6. a2 - b2 = (a + b)(a - b)
7. a3 - b3 = (a - b)(a2 + ab + b2)
8. a3 + b3 = (a + b)(a2 - ab + b2)
"""


# Create a message
messages = [
    ("system", "You are a equation formatter assistant."),
    ("human", "fix the formate of the eqution given bellow, the text is extracted from pdf so formate of the equation is lost. fix it. {identities}".format(identities=identities))
]

# Invoke the model
response = geminiChatModel.invoke(messages)

print(response.content)
```
