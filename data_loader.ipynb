```python
import os
from dotenv import load_dotenv, find_dotenv
from langchain_groq import ChatGroq

_  = load_dotenv(find_dotenv())

chat_model = ChatGroq(model="llama-3.1-8b-instant")
```

```python
messages = [
    ("system", "you are a helpfull assitant"),
    ("human", "tell me about gpt 5 model")
]

response = chat_model.invoke(messages)

print(response.content)
```

# Data Loader

```python
# text data loader

from langchain_community.document_loaders import TextLoader

loader = TextLoader("text.txt")
loaded_data = loader.load()

print(loaded_data)

print(loaded_data[0].page_content)
```

```python
# csv loader
from langchain_community.document_loaders import CSVLoader

loader = CSVLoader("csv.csv")
loaded_data = loader.load()

print(loaded_data)

for data in loaded_data:
    print(data.page_content)
```

```python
# wikipidia data loader
from langchain_community.document_loaders import WikipediaLoader

loader = WikipediaLoader(query="gpt 5", load_max_docs=1)

loaded_data = loader.load()[0].page_content

loaded_data
```

```python
from langchain_core.prompts import ChatPromptTemplate

chat_template = ChatPromptTemplate.from_messages(
    [
        ("human", "answer this {question}, here is some context {context}")
    ]
)

messages = chat_template.format_messages(
    question = "tell me about gpt 5",
    context = loaded_data
)

response = chat_model.invoke(messages)

response.content
```
