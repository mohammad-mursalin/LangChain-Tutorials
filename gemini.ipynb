```python
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
import os

load_dotenv()  # loads GEMINI_API_KEY

# Create a Gemini 2.0 Flash model
geminiChatModel = ChatGoogleGenerativeAI(
    model="gemini-2.0-flash",
    api_key=os.getenv("GEMINI_API_KEY")
)

# Create a message
messages = [
    ("system", "You are a programming assistant."),
    ("human", "Write a Java Spring Boot controller for JWT authentication.")
]

# Invoke the model
response = geminiChatModel.invoke(messages)

print(response.content)
```
