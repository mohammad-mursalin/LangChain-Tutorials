```python
from langchain_community.document_loaders import TextLoader

loaded_document = TextLoader("text.txt").load()
print(loaded_document[0].page_content)
```

# Character Text Splitter

```python
from langchain_text_splitters import CharacterTextSplitter

text_splitter = CharacterTextSplitter(
    separator = "\n",
    chunk_size = 100,
    chunk_overlap = 10,
    length_function = len,
    is_separator_regex=False
)
```

```python
splitted_text = text_splitter.create_documents([loaded_document[0].page_content])
```

```python
splitted_text
```

```python
len(splitted_text)
```

```python
splitted_text[0].page_content
```

# Recursive Character Text Splitter

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 200,
    chunk_overlap = 50
)
```

```python
splitted_text = text_splitter.split_text(loaded_document[0].page_content)

splitted_text
```
