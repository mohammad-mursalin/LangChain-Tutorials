```python
import os
from dotenv import load_dotenv, find_dotenv
from langchain_groq import ChatGroq

_ = load_dotenv(find_dotenv())

model = ChatGroq(model = "llama-3.1-8b-instant")
```

```python
model.invoke("Tell me the latest news of India and pakistan war in 2025")
```

```python
from langchain_community.tools.tavily_search import TavilySearchResults
search = TavilySearchResults(max_results=3)
search.invoke("Tell me the latest news of India and pakistan war in 2025")
```

```python
tools = [search]
```

```python
from langgraph.prebuilt import create_react_agent

agent = create_react_agent(model, tools)
```

```python
from langchain_core.messages import HumanMessage

response = agent.invoke({"messages": [HumanMessage(content="Tell me the lastest news of India and pakistan war in 2025")]})

response
```

```python
for chunk in agent.stream(
    {"messages": [HumanMessage(content="Tell me the lastest news of India and pakistan war in 2025")]}
):
    print(chunk)
    print("----")
```
